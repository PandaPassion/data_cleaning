```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
df = pd.read_csv('/Users/bujoriustin/Desktop/temp/healthcare-dataset-stroke-data.csv')
```

```python
df.ever_married.replace({'Yes': 0, "No": 1}, inplace = True)
df.gender.replace({'Male': 0, "Female": 1}, inplace = True)
df.work_type.replace({'Self-employed': 0, 'Private' : 1, 'Govt_job': 2, 'children':3, 'Never_worked':4}, inplace = True)
df.Residence_type.replace({'Urban': 0, "Rural": 1}, inplace = True)
```

```python
df.gender.replace('Other', 2, inplace = True)
```

```python
df.smoking_status.replace({'formerly smoked': 1, 'never smoked': 0, 'smokes':3, 'Unknown': None}, inplace = True)
```

```python
imputer = KNNImputer(n_neighbors=5)
imputer.fit(df)
trans = imputer.transform(df)
trans = pd.DataFrame(trans, columns = ['id', 'gender', 'age', 'hypertension', 'heart_disease', 'ever_married',
       'work_type', 'Residence_type', 'avg_glucose_level', 'bmi',
       'smoking_status', 'stroke'])
```
